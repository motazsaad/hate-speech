```
def preprocess(text):
  text = re.sub("RT @USER:", " ", str(text))
  text = re.sub("@USER", " ", str(text))
  text = re.sub("URL", " ", str(text))
  text = re.sub("<LF>", " ", str(text))
  tokens = []
  for token in text.split():
    # if token not in stop_words:
      tokens.append(token)
  return " ".join(tokens)
```
